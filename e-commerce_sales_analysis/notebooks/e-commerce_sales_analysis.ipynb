```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("E-Commerce.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

# Converting date columns

```python
df["Order Date"]=pd.to_datetime(df["Order Date"], errors = "coerce")
df["Ship Date"]=pd.to_datetime(df["Ship Date"], errors = "coerce")
```

```python
df.info()
```

```python
df.head()
```

```python
df["Order Month"] = df["Order Date"].dt.month
df["Order Year"] = df["Order Date"].dt.year
df["Order Day of Week"] = df["Order Date"].dt.dayofweek
```

```python
df.head()
```

# Monthly Sales Analysis

```python
monthly_sales = df.groupby("Order Month")["Sales"].sum().reset_index().sort_values("Order Month")
monthly_sales 
```

```python
plt.figure(figsize=(10,5))
sns.barplot(
    data = monthly_sales,
    x = "Order Month",
    y = "Sales",
    edgecolor="black",
    linewidth=1,
    color="lightskyblue"
)
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Monthly Sales of the Store")
plt.show()
```

# Sales by Category

```python
category_sales = df.groupby("Category")["Sales"].sum().reset_index().sort_values("Sales", ascending=False)
category_sales 
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data = category_sales,
    x = "Category",
    y = "Sales", edgecolor="black",
    linewidth=1,
    color="indianred"
)
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.title("Sales by Product Category")
plt.show()
```

# Sales analysis by Sub Category

```python
subcategory_sales = df.groupby("Sub-Category")["Sales"].sum().reset_index().sort_values("Sales").reset_index(drop=True)
#subcategory_sales = df.groupby("Sub-Category")["Sales"].sum().reset_index().sort_values("Sales") it will show index value randomly
subcategory_sales 
```

```python
plt.figure(figsize=(10, 12))
sns.barplot(data=subcategory_sales,
            y="Sub-Category", 
            x="Sales",
            edgecolor="black",
            linewidth=1,
            color="lightskyblue"
           
)
plt.xlabel("Total Sales")
plt.ylabel("Sub-Category")
plt.title("Sales by Sub-Category")
plt.show()
```

# sales and profit - customer segment

```python
segment_summary = df.groupby("Segment")[["Sales", "Profit"]].sum().reset_index()
segment_summary
```

```python
segment_summary.set_index("Segment")[["Sales", "Profit"]].plot(
    kind="bar",
    figsize=(9,5),
    edgecolor="black",
    linewidth=1,
)

plt.xlabel("Customer Segment")
plt.ylabel("Amount")
plt.title("Sales and Profit by Customer Segment")
plt.xticks(rotation=0)  
plt.ticklabel_format(style='plain', axis='y')
#Show the y-axis numbers in normal form, not scientific notation.So it removes 1e6, 1e5, etc. from the axis.
# plain means Normal numbers (1000000) 
plt.show()
```

# sales to profit ratio

```python
sales_profit_by_segment = df.groupby("Segment")[["Sales", "Profit"]].sum().reset_index()

sales_profit_by_segment["Sales_Profit_Ratio"] = sales_profit_by_segment["Profit"]/sales_profit_by_segment["Sales"]
sales_profit_by_segment[["Segment", "Sales_Profit_Ratio"]]-
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=sales_profit_by_segment,
    x="Segment",
    y="Sales_Profit_Ratio",
    edgecolor="black",
    linewidth=1,
    color="lightskyblue"
)

plt.xlabel("Customer Segment")
plt.ylabel("Sales to Profit Ratio")
plt.title("Sales to Profit Ratio by Customer Segment")
plt.show()
```

# Year-over-Year (YoY) Sales Growth Analysis

```python
yearly_sales = (df.groupby("Order Year")["Sales"].sum().reset_index().sort_values("Order Year"))

yearly_sales
```

```python
yearly_sales["YoY_Growth_%"]= yearly_sales["Sales"].pct_change() * 100

yearly_sales
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=yearly_sales.dropna(),
    x="Order Year",
    y="YoY_Growth_%",
    edgecolor="black",
    linewidth=1,
    color="brown"
)

plt.xlabel("Order Year")
plt.ylabel("YoY Growth (%)")
plt.title("Year-over-Year Sales Growth (%)")
plt.show()
```

# Seasonality Detection

```python
# Sales by Year and Month -
ym_sales = (df.groupby(["Order Year", "Order Month"])["Sales"].sum().reset_index())
ym_sales
```

```python
heat_data = ym_sales.pivot(index="Order Year", columns="Order Month", values="Sales")

plt.figure(figsize=(12, 5))
sns.heatmap(heat_data, annot=False, cmap="coolwarm")
plt.title("Seasonality Heatmap: Sales by Year and Month")
plt.xlabel("Month")
plt.ylabel("Year")
plt.show()
```

# Top 5 and Bottom 5 States by Profit

```python
state_profit = df.groupby("State")["Profit"].sum().reset_index().sort_values("Profit", ascending=False)

top5 = state_profit.head(5)
bottom5 = state_profit.tail(5)

top5, bottom5
```

```python
plt.figure(figsize=(12,5))

# Top 5 States
plt.subplot(1, 2, 1)
sns.barplot(
    data=top5,
    x="Profit",
    y="State",
    edgecolor="black",
    linewidth=1,
    color="lightskyblue"
)
plt.title("Top 5 States by Total Profit")
plt.xlabel("Total Profit")
plt.ylabel("State")

# Bottom 5 States
plt.subplot(1, 2, 2)
sns.barplot(
    data=bottom5,
    x="Profit",
    y="State",
    edgecolor="black",
    linewidth=1,
    color="brown"
)
plt.title("Bottom 5 States by Total Profit")
plt.xlabel("Total Profit")
plt.ylabel("State")

plt.tight_layout()
plt.show()
```

# Loss-Making Cities Analysis

```python
city_profit = df.groupby("City")["Profit"].sum().reset_index()

loss_cities = city_profit.query("Profit < 0").sort_values("Profit")

loss_cities.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    data=loss_cities.head(),
    x="Profit",
    y="City",
    edgecolor="black",
    linewidth=2,
    color="brown"
)

plt.title("Top 5 Loss-Making Cities by Total Profit")
plt.xlabel("Total Profit")
plt.ylabel("City")
plt.show()
```

# High-Sales but Low-Profit Products

```python
product_summary = (df.groupby("Product Name")[["Sales", "Profit"]].sum().reset_index())

product_summary 
```

```python
# Sort by Sales (highest first)
product_summary = product_summary.sort_values("Sales", ascending=False)

# Take top 20 selling products
top_sales_products = product_summary.head(20)

# Use query() to get low or negative profit products
high_sales_low_profit = top_sales_products.query("Profit <= 0")

high_sales_low_profit
```

```python
plt.figure(figsize=(9,5))
sns.barplot(
    data=high_sales_low_profit,
    x="Profit",
    y="Product Name",
    edgecolor="black",
    linewidth=1,
    color="lightskyblue"
)
plt.title("High-Sales but Low-Profit Products")
plt.xlabel("Profit")
plt.ylabel("Product Name")
plt.show()
```
