```python
df["Release_Date"] = pd.to_datetime(df["Release_Date"])

print(df["Release_Date"].dtypes)
```

```python
df["Year"] = df["Release_Date"].dt.year
df["Month"] = df["Release_Date"].dt.month
```

# Dropping the columns

```python
cols = ["Overview", "Poster_Url"]

df.drop(cols, axis=1, inplace=True)
```

```python
df.columns
```

```python
df.head()
```

Categorizing Vote_Average column

We would cut the Vote_Average values and make 4 categories : popular average below_avg not_popular to describe it more using categorize_col()
function provided above

```python
#df['Vote_Average'] = pd.to_numeric(df['Vote_Average'], errors='coerce')
#  when Column is object (string),category, mixed types then use this line. It converts the column Vote_Average into numeric type (float).{"7.5" to 7.5}

labels = ['not_popular', 'below_avg', 'average', 'popular']

df['Vote_Average'] = pd.qcut(
    df['Vote_Average'],
    q=4,
    labels=labels,
    duplicates='drop'
)
```

```python
df
```

```python
df['Vote_Average'].value_counts()
```

```python
df.dropna(inplace = True)

df.isna().sum()
```

# We would split genres into a list and then explode our dataframe to have only one genre per row for each movie

```python
df['Genre'] = df['Genre'].str.split(', ')

df = df.explode('Genre').reset_index(drop=True)
```

```python
df.head()
```

```python
# Casting column into category

df['Genre'] = df['Genre'].astype('category')

df['Genre'].dtypes
```

```python
df.info() 
```

```python
df.nunique()
```

```python
df.head()
```

# What is the most frequent genre of movies released on Netflix?

```python
df['Genre'].describe()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(
    y='Genre',
    data=df,
    order=df['Genre'].value_counts().index,
    edgecolor="black",
    linewidth=1,
    color="lightskyblue"
    
)

plt.title("Most Frequent Genres on Netflix")
plt.xlabel("Number of Movies")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

```python
df.head()
```

# Which Vote_Average category has highest total votes

```python
vote_category_votes = (df.groupby('Vote_Average')['Vote_Count'].sum().sort_values(ascending=False))

vote_category_votes
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x=vote_category_votes.values,
    y=vote_category_votes.index,
    edgecolor="black",
    linewidth=1,
    color="lightskyblue"
)

plt.title("Total Votes by Vote_Average Category")
plt.xlabel("Total Vote Count")
plt.ylabel("Vote_Average Category")
plt.ticklabel_format(style='plain', axis='x')

plt.show()
```

# Which Vote_Average category contains the highest number of movies

# Which year has the most filmed movies

```python
top_year = df['Year'].value_counts().idxmax()
# idxmax() -> Returns the index (year) that has the highest count.

print("Year with most movies:", top_year)
```

```python
plt.figure(figsize=(12,6))

sns.histplot(df['Year'],
             bins=30,
             edgecolor='black',
             linewidth=1,
             color="lightskyblue"
)

plt.title("Movies Released per Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")

plt.show()
```
